#### Nivel 1 

*  1 - Leia um número inteiro e informe se ele é positivo, negativo ou zero.

```python
# Exercicio 1

numero = int(input("Digite um numero"))

if numero > 0:
    print("positivo")
else:
    print("negativo")
```

```python
#  Exercicio 2

idade = int(input("Digite sua idade"))

if idade > 0 and idade <=12:
    print("Criança")
elif idade >= 13 and idade <=17:
    print("Adolecente")
elif idade >=18:
    print("Adulto")
else: 
    print("Invalido")
```

```python
# Exercicio 3
n1 = int(input("Digite um numero"))
n2 = int(input("Digite um numero"))

if n1 > n2:
    print(f"{n1} é maior")
elif n2 > n1:
    print(f"{n2} é maior")
else: 
    print("São iguais")
```

```python
## Exercicio 5

numero = int(input("Digite um numero"))

if numero %2 == 0:
    print(f"{numero} é par")
else:
    print(f"{numero} é impar")
```

```python
## Exercicio 6

valor_produto = float(input("Digite o preço do produto"))

if valor_produto <= 100:
    desconto =  valor_produto * 0.05
    print(f"Aplicado 5% de desconto no valor {valor_produto}\nPreço final { valor_produto - desconto}", )
elif valor_produto >=100.01 and valor_produto <= 300:
    desconto =  valor_produto * 0.10
    print(f"Aplicado 10% de desconto no valor {valor_produto}\nPreço final {valor_produto - desconto}", )
else:
    desconto = valor_produto * 0.15
    print(f"Aplicado 15% de desconto no valor {valor_produto}\nPreço final {valor_produto - desconto}", )
```

```python
# Exercicio 7

lado1 = int(input('Digite um valor para o lado 1: '))
lado2 = int(input('Digite um valor para o lado 2: '))
lado3 = int(input('Digite um valor para o lado 3: '))

#verificando se é ou não um triângulo
if lado1 + lado2 > lado3 and lado2 + lado3 > lado2 and lado2 + lado3 > lado1:
    #descobrindo o tipo de triângulo
    if lado1 == lado2 == lado3:
        print('Equilátero.')
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print('Isósceles.')
    else:
        print('Escaleno.')
else:
    print('Com os valores passados, não é possível formar um triângulo.')
```

```python
# Exercicio 8
numero = int(input('Digite um número: '))

#loop de 1 até o número lido
for i in range(1, numero + 1):
    print(i)
```

```python
# Exercicio 9 

numero = int(input('Digite um número: '))

#loop de 1 até o número lido em ordem decrescente
for i in range(numero, -1, -1):
    print(i)
```

```python
# Exercicio 10

numero = int(input('Digite um número inteiro: '))

print(f'Tabuada do {numero}:')
for i in range(1, 11):
    print(f'{numero} X {i}: {numero * i}')
```
