7. Leia 3 lados e verifique se formam triângulo. Se formar, classifique: equilátero, isósceles ou
escaleno.

```python
# Exercicio 7

lado1 = int(input('Digite um valor para o lado 1: '))
lado2 = int(input('Digite um valor para o lado 2: '))
lado3 = int(input('Digite um valor para o lado 3: '))

#verificando se é ou não um triângulo
if lado1 + lado2 > lado3 and lado2 + lado3 > lado2 and lado2 + lado3 > lado1:
    #descobrindo o tipo de triângulo
    if lado1 == lado2 == lado3:
        print('Equilátero.')
    elif lado1 == lado2 or lado1 == lado3 or lado2 == lado3:
        print('Isósceles.')
    else:
        print('Escaleno.')
else:
    print('Com os valores passados, não é possível formar um triângulo.')
```

22. Leia uma lista de N números (N informado pelo usuário) e calcule a média apenas dos
números entre 0 e 100 (ignore os demais).

```python
#Exercicio 22
lista = []
n = int(input("Digite quantos numeros quer adicionar a lista"))
soma = 0
contador = 0

for i in range(n):
    numero = int(input("Digite numeros entre 0 a 100: "))

    if numero >= 0  and numero <=100:
        soma+= numero
        contador+=1
        lista.append(numero)
    else:
        print("Digite um numero valido")
        break
if contador >0:
    media = soma/contador
    print(f"Soma:{soma}\nMedia:{media:.2f}")
else:
    print("Média não encontrada")
```

23. Leia N números e determine o segundo maior valor (considere que pode haver repetidos).


```python
#Exercicio 23
lista = []
n = int(input("Digite quantos numeros para adicionar na lista"))

for i in range(n):
    numero = int(input("Digite um numero"))
    lista.append(numero)

    maior = 0
    segundo = 0
    for i in lista:
        if i > maior:
            maior = i
            for j in lista:
                if j > segundo  and j < maior:
                    segundo = j

                
print("Lista:", lista)                    
print("Maior:", maior)
print("Segundo maior:", segundo)



```

28. Leia um número N e verifique se ele é um número perfeito (soma dos divisores próprios =
N).

```python
#Exercicio 28

numero = int(input("Digite um numero"))

soma = 0

for i in range(1,numero):
    if numero % i == 0:
        soma += i
if soma == numero:
    print(numero ,"É um numero perfeito")
else:
    print(numero,"Não é um numero perfeito")
```

29. Leia uma palavra e verifique se ela é palíndromo (lê igual de trás para frente).

```python
palavra = str(input("Digite um palindromo"))

if palavra == palavra[::-1]:
    print(palavra, "É um palindromo")
else:
    print(palavra, "Não é um palindromo")
```

30. Peça um número N e imprima os N primeiros termos da sequência de Fibonacci (começando em 0, 1).


```python
fibonacci = int(input("Digite um numero"))

anterior = 0
proximo = 1

for i in range(fibonacci):
    sequencia = anterior + proximo
    anterior = proximo
    proximo = sequencia
    print(anterior, "+", proximo)
print(f" Primieros termos da sequencia de {sequencia} numeros")
```
