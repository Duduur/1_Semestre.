### 4. Pesquisa Linear (retornar índice)
Enunciado: Crie uma função que faça pesquisa linear em uma lista e retorne o índice
do elemento se encontrado ou -1 se não estiver presente

```python
lista = [2,3,4,5,62,52]

def pesquisa_linear(lista, elemento):
    for i in range(len(lista)):
        if lista[i] == elemento:
            return f"Indice do {elemento} = ",i
    return -1
print(pesquisa_linear(lista, 4))
```

### 5. Exclusão de Elemento (sem usar remove)
Enunciado: Crie uma função que remova a primeira ocorrência de um elemento de
uma lista sem utilizar o método remove.

```python
lista = [2,3,4,5,62,52]
def exclusao(lista,elemento):
    for i in range(len(lista)):
        if lista[i] == elemento:
            del lista[i]
            break
    return lista
print(exclusao(lista, 62))
    
```

### 6. Inserção em Lista Ordenada
Enunciado: Crie uma função que insira um elemento em uma lista ordenada,
mantendo a ordem sem usar sort.


```python
lista = [2,3,4,5,62,52]

def inserir_ordem(lista, elemento):
    posicao = 0
    while posicao < len(lista) and lista[posicao] < elemento:
        posicao += 1
    lista.insert(posicao, elemento)
    return lista

print(inserir_ordem(lista, 78))
```

### 14. Calcular o Quadrado de Cada Elemento
Enunciado: Crie uma função que retorne uma nova lista com o quadrado de cada
elemento da lista original.

```python
vetor = [2,3,4,5,62,52,3,2,157,67,67,67]

def quadrado(vetor):
    nova_lista = []
    contador = 0
    for i in vetor:
        aoquadrado = i**2
        nova_lista.append(aoquadrado)
        contador+=1
    return nova_lista
print(quadrado(vetor))
```

### 16. Remover Duplicados
Enunciado: Crie uma função que receba uma lista e retorne uma nova lista com os
elementos duplicados removidos, mantendo a ordem original.

```python
vetor = [2,3,4,5,62,52,3,2,157,67,67,67]

def remocao_duplicatas(vetor):
    nova_lista = []

    for i in vetor:
        contador = 0
        for j in nova_lista:
            if i == j:
                contador +=1

        if contador == 0:
            nova_lista.append(i)

    return "Lista sem duplicados", nova_lista

print(remocao_duplicatas(vetor))

        
        
```

### 17. Inserir Elemento se Não Existir

não estiver presente.

```python
vetor = [2,3,4,5,62,52,3,2,157,67,67,67]


def inserindo_novos(vetor,elemento):
    contador =0
    for i in vetor:
        if i == elemento:
            contador+=1

    if contador == 0:
        vetor.append(elemento)
    else:
        return False

    return "Lista:", vetor

print(inserindo_novos(vetor,1))
```

### 18. Combinar Duas Listas
Enunciado: Crie uma função que combine duas listas e retorne uma única lista com
todos os elementos.


```python
vetor = [2,3,4,5,62,52,3,2,157,67,67,67]
vetor2 = [525, 20777, 34250]

def combinar_listas(lista1, lista2):
    nova_lista = []

    for i in lista1:
        nova_lista.append(i)

    for j in lista2:
        nova_lista.append(j)

    return nova_lista

print(combinar_listas(vetor, vetor2))
```

### 20. Calcular Média
